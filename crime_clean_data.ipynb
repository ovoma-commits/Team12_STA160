```python
import pandas as pd
housing = pd.read_csv("combined_housing_data.csv")
crime = pd.read_excel("violent-crime-rate-california-2000-2013.xlsx")

housing.head()
crime.head()
```

```python
#clean
crime.columns
```

```python
crime_clean = crime[["reportyear", "county_name", "rate"]].copy()

crime_clean.rename(columns={
    "reportyear": "year",
    "county_name": "county",
    "rate": "violent_crime_rate"
}, inplace=True)

crime_clean = crime_clean.dropna(subset=["county", "violent_crime_rate"]).drop_duplicates()


crime_clean["county_clean"] = (
    crime_clean["county"].astype(str)
    .str.strip().str.lower()
    .str.replace(" county", "", regex=False)
)

crime_clean.head()
```

```python
crime_latest = (
    crime_clean.groupby("county_clean")["violent_crime_rate"]
    .mean().reset_index()
)
crime_latest["violent_crime_rate_percent"] = (crime_latest["violent_crime_rate"] / 1000) * 100
crime_latest["violent_crime_rate_percent"] = crime_latest["violent_crime_rate_percent"].round(2)
crime_latest = crime_latest.drop(columns=["violent_crime_rate"])

crime_latest.head()
```

```python

crime_export = crime_latest[["county_clean", "violent_crime_rate_percent"]].copy()

crime_export.to_csv("crime_cleaned_percent.csv", index=False)

from google.colab import files
files.download("crime_cleaned_percent.csv")
```
